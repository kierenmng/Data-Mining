```python
#separate string using seps, removes bs from beginning
def eval_string_from_file(file):
    arr = []
    with open(file,"r", encoding = 'utf-8') as file:
        for line in file:
            result = line[46:]
            result2 = result.replace("<SEP>", " ")
            result3 = result2.replace("\n","")
            arr.append(result3)
            #print(result2)
    file.close()
    return arr

def remove_parens_contents(string):
#     new = string.split('(', 1)[0]
    n = len(string)
    while n > 0:
        n= n-1
        if(string[n] == '('):
            break
    new = string[:n]
    return new
```
